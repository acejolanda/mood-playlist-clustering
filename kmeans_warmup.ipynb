```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simple 2D dataset
X = np.array([  
                [0, 0], [12, 1], [1, 2], [9, 4], [1, 0], [2, 3],
                [5, 2], [3, 3], [4, 2], [12, 2], [1, 3], [10, 0]
            ])
```

```python
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto")
kmeans.fit(X) # method takes data and performs the clustering process
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Cluster Labels:", labels) 
print("Cluster Centers:", centroids) 
```

```python
# 1st visualization

plt.figure(figsize=(8, 6))
colors = ['blue', 'orange']

cluster_labels = [0, 1]

for cluster in cluster_labels:
    plt.scatter(
        X[labels == cluster, 0], X[labels == cluster, 1], 
        s=50, c=colors[cluster], label=f"Cluster {cluster}"
    )
    plt.scatter(
        centroids[cluster, 0], centroids[cluster, 1],
        s=300, c=colors[cluster], marker='o', 
        label=f"Centroid {cluster}"
    )
plt.title("Original Data Points with centroids", fontsize=14)
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.legend(loc = "upper left", fontsize = 8.5)
plt.grid(alpha=0.3)
plt.show()
```

```python
new_arr = np.array([
        [0, 1], [10, 5], [2, 4]
        ])
```

```python
predicted_labels = kmeans.predict(new_arr)
print("Predicted Labels for new points:", predicted_labels)
```

```python
# 2nd visualization

plt.figure(figsize=(8, 6))
for cluster in cluster_labels:
    plt.scatter(
        X[labels == cluster, 0], X[labels == cluster, 1], 
        s=50, c=colors[cluster], label=f"Cluster {cluster}"
    )
    plt.scatter(
        centroids[cluster, 0], centroids[cluster, 1],
        s=300, c=colors[cluster], marker='o', 
        label=f"Centroid {cluster}"
    )
plt.scatter(
    new_arr[:, 0], new_arr[:, 1], 
    s=200, c=[colors[label] for label in predicted_labels], 
    marker='*', label="New Points"
)
plt.title("All data points (original + centroids + new)", fontsize=14)
plt.xlabel("Feature 1", fontsize=12)
plt.ylabel("Feature 2", fontsize=12)
plt.legend(loc = "upper left", fontsize = 8.5)
plt.grid(alpha=0.3)
plt.show()
```
